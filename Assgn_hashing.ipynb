# Hashing Algorithm

Hashing Vectorizer converts text into fixed-length numerical feature vectors using a hash function, eliminating the need to build and store a vocabulary. It is memory-efficient and well-suited for large-scale text processing.

```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
from sklearn.metrics import classification_report
```

```python
dataset = pd.read_csv("news.csv").drop(columns=["Unnamed: 0"])
```

```python
dataset
```

```python
y=dataset["label"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(dataset['text'], y, test_size=0.33, random_state=53)
```

Here we no need to call fit() as it is not learning any vocabulary from the training data here. It simply hashes words directly.

```python
hash_vectorizer = HashingVectorizer(stop_words='english',n_features=2**18,alternate_sign=False)
hash_train = hash_vectorizer.transform(X_train)
print(hash_train)
hash_test = hash_vectorizer.transform(X_test)
```

Every word is converted into a large integer using a mathematical formula.
word -- hash function -- modulo operation with no of columns(no. of features) -- if remainder is 0 it belongs to col 0.
feature_index = hash(word) % n_features

n_features = 2**18
each document has 262,144 columns.

A larger value reduces the chance that different words map to the same feature.

```python
clf = MultinomialNB()

clf.fit(hash_train, y_train)
pred = clf.predict(hash_test)
score = metrics.accuracy_score(y_test, pred)
print("accuracy:   %0.3f" % score)

cm = metrics.confusion_matrix(y_test, pred, labels=['FAKE', 'REAL'])
```

```python
from sklearn.metrics import classification_report

report=classification_report(y_test, pred)
```
