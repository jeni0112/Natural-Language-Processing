# LDA

Latent Dirichlet Allocation (LDA) is a topic modeling technique that automatically identifies hidden topics in a collection of text documents. It groups related words into topics and estimates the probability of each topic being present in a document.

```python
import pandas as pd
```

```python
df = pd.read_csv("Mahas_Group.txt",header=None,on_bad_lines="skip",encoding='utf8')
```

```python
df
```

```python
df=df.drop(0)
df.columns=['Date','Chat']
Message=df["Chat"].str.split("-",n=1,expand=True)
df["Time"]=Message[0]
Message1=Message[1].str.split(":",n=1,expand=True)
df["Name"]=Message1[0]
df["Chat"]=Message1[1]
df=df[["Date","Time","Name","Chat"]]
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

 Create TF-IDF Document-Term Matrix

```python
tfidf=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
dtm=tfidf.fit_transform(df["Chat"])
```

Create LDA model

```python
from sklearn.decomposition import LatentDirichletAllocation
lda_model = LatentDirichletAllocation(n_components=5,random_state=42)
lda_model.fit(dtm)
```

Display top 10 words for each topic

```python
for index, topic in enumerate(lda_model.components_):
    results = [tfidf.get_feature_names_out()[i] for i in topic.argsort()[-10:]]
    print(results)
```

Get topic probabilities for each document

```python
topic_results = lda_model.transform(dtm)
```

Assign the dominant topic

```python
df = df.copy()
df["Topic"] = topic_results.argmax(axis=1)
```
